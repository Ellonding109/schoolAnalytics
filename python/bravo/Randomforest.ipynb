```python
import numpy as np 
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

irisdata = load_iris()
print(irisdata)
```

## get the info on the dataset

```python

df = pd.DataFrame(irisdata.data, columns = irisdata.feature_names)

print(df.describe())
display(df.head(10))
print(df.info())
# GHECK FOR ANY NULL VALUES 
print(df.isnull().sum())
```

```python
# create the data to be predicted as a value of words 

df['species'] = pd.Categorical.from_codes(irisdata.target, irisdata.target_names)
df
```

```python
# splitting the dataset for the purpose of training and testing

df['istrain'] = np.random.normal(0,1,len(irisdata.data))<=.75
display(df)
# create the split 
train, test = df[df['istrain']== True], df[df['istrain']==False]
print(len(test))
y = 20
y<=10
```

```python
features = df.columns[:4]
features
```

```python
# make the species to be understood by the comp converting the species names to digits
y = pd.factorize(train['species'])[0]
print(y)
len(y)
```

```python
# import random forest for the calassification 
from sklearn.ensemble import RandomForestClassifier
# define the classifier

fcl = RandomForestClassifier(n_jobs=2, random_state = 0 )
```

```python
# define the features to specifie the training columns
features = df.columns[:4]
print(len(train[features]))
display(train[features])
fcl.fit(train[features],y)
```

```python
fcl.predict(test[features])
# show the features in the test
print(len(test[features]))
display(test[features])
# pd.factorize(test[features])
```

```python
# the tree probability 

fcl.predict_proba(test[features])
```

```python
# now to human readable
predicted =irisdata.target_names[fcl.predict(test[features])]
display(predicted)
print(irisdata.target_names)
```

```python
matchTable = pd.crosstab(test['species'],predicted, rownames = ['actual'],colnames= ["predicted"])
matchTable
```
